```python
soup=BeautifulSoup(resp.text,"html.parser")
soup
```

```python
reviews = soup.find_all("div", {"class": "a-section review aok-relative"})
len(reviews)
```

```python
def get_review_stars(soup_obj: BeautifulSoup) -> str:
    try:
        review_str = soup_obj.find('a', {"class": "a-link-normal"}).get_text()
        return review_str
    except Exception as e:
        return 'no_stars'
        print(e)

def get_customer_name(soup_obj: BeautifulSoup) -> str:
    try:
        review_name = soup_obj.find('span', {"class": "a-profile-name"}).get_text().strip()
        return review_name
    except Exception as e:
        return 'no_name'
        print(e)
        
def get_review_title(soup_obj:BeautifulSoup) -> str:
    try:
        review_title=soup_obj.find('a',{'data-hook':'review-title'}).get_text().strip()
        return review_title
    except Exception as e:
        return 'no_title'
        print(e)

def get_review_body(soup_obj: BeautifulSoup) -> str:
    try:
        review_body = soup_obj.find('span', 
            {"class": "a-size-base review-text review-text-content"}
        ).get_text().strip()
        return review_body
    except Exception as e:
        return 'no_body'
        print(e)
```

```python
for r in reviews:
    print(r.find('a',{'data-hook':'review-title'}).get_text())
    print()
```

```python
def get_review_components(soup_obj: BeautifulSoup) -> dict:
    return {
        'stars': get_review_stars(soup_obj), 
        'body': get_review_body(soup_obj), 
        'name': get_customer_name(soup_obj),
        'title': get_review_title(soup_obj)
    }
```

```python
records = [get_review_components(rev) for rev in reviews]
records
```

```python
df = pd.DataFrame.from_records(records)
df
```

```python
ALL_REVIEWS = []

for u in list_url:
    print(u)
    resp = requests.get(u, headers=headers)
    soup = BeautifulSoup(resp.text, "html.parser")
    reviews = soup.find_all("div", {"class": "a-section review aok-relative"})
    records = [get_review_components(rev) for rev in reviews]
    ALL_REVIEWS.extend(records) 
    
    
op = pd.DataFrame.from_records(ALL_REVIEWS)
op
```

# Text Mining 

```python
df1=pd.read_csv('echo.csv')
df1
```

```python
df1.pop('Unnamed: 0')
```

```python
df1.pop('title')
```

```python
df1
```

```python
df1['stars'].replace({'3.0 out of 5 stars':3,'2.0 out of 5 stars':2,'1.0 out of 5 stars':1,'4.0 out of 5 stars':4,'5.0 out of 5 stars':5},inplace=True)
```

```python
df1
```

```python
df1 = df1.rename(columns={"body":"review"})
```

```python
df1
```

```python
first_column = df1.pop('name')
df1.insert(0, 'name', first_column)

third_column=df1.pop('stars')
df1.insert(2,'stars',third_column)
```

```python
df1
```

```python
df1.isna().sum()
```

```python
len(df1)
```

```python
for i in range(0,len(df1)-1):
    if type(df1.iloc[i]['review']) != str:
        df1.iloc[i]['review'] = str(df1.iloc[i]['review'])
```

```python
df1 = df1[df1['stars'] != 3]
```

```python
def sentiment(n):
    return 1 if n >= 4 else 0
df1['sentiment'] = df1['stars'].apply(sentiment)
df1.head()
```

```python
df1.stars.unique()
```

```python
df1.shape
```

```python
def combined_features(row):
    return row['name'] + ' '+ row['review']
df1['all_features'] = df1.apply(combined_features, axis=1)
df1.head()
```

```python
X = df1['all_features']
y = df1['sentiment']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
ctmTr = cv.fit_transform(X_train)
X_test_dtm = cv.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(ctmTr, y_train)
```

```python
y_pred_class = model.predict(X_test_dtm)
y_pred_class
```

```python
accuracy_score(y_test, y_pred_class)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred_class))
```

```python
print(confusion_matrix(y_test,y_pred_class))
```

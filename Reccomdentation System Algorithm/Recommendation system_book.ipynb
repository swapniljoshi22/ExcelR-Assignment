```python

book_df= df.pivot_table(index='user_id',columns='Book_',values='rating').reset_index(drop=True)
book_df
```

```python

book_df.index= df.user_id.unique()
book_df
```

```python
#Impute those NaNs with 0 values
book_df.fillna(0, inplace=True)
book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book_df.values,metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index=df.user_id.unique()
user_sim_df.columns=df.user_id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
#fill diagonal value == 0
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:5,0:5]
```

```python
#most similar user 
user_sim_df.idxmax(axis=1)
```
